```python
from scipy.optimize import linprog
```

```python
# Serviceeinheit 1
```

```python
obj = [-300, -600, 0, 0] # da linprog nur minimiert gilt folgende Rechenregel für die Zielfunktion: max(f(x))=-min(-f(x))
```

```python
lhs_ineq = [[ 700,  300, -40, -500],  # BB1
            [ 300,  600, -50, -500],  # BB2
            [ 200,  700, -50, -400],  # BB3
            [ 400,  600, -50, -500],  # BB4
            [ 500,  400, -40, -400],  # BB5
            [ 500,  500, -50, -500],  # BB6
            [ 800,  500, -40, -600],  # BB7
            [ 300,  200, -30, -400]]  # BB8
```

```python
rhs_ineq = [0, 0, 0, 0, 0, 0, 0, 0]
```

```python
lhs_eq = [[0, 0, 50, 500]] 
rhs_eq = [1]
```

```python
bnd = [(0, float("inf")),  # u1
       (0, float("inf")),  # u2
       (0, float("inf")),  # v1
       (0, float("inf"))]  # v2
```

```python
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              A_eq=lhs_eq, b_eq=rhs_eq, bounds=bnd,
              method="revised simplex")
```

```python
opt
```

```python
print("Effizienz Serviceeinheit 1: {}".format(-1*opt.fun))
print("Nutzengewichte Arbeitsstunde und Materialeinsatz: {}".format(opt.x[1:]))

effizienz_bb2 = -1 * opt.fun
print(f"Effizienz Busbahnhof 2: {effizienz_bb2:.4f}") # Gibt dir die 87,39% (0.8739)
```

```python
obj = [-400, -600, 0, 0]
```

```python
lhs_ineq = [[ 700,  300, -40, -500],  # BB1
            [ 300,  600, -50, -500],  # BB2
            [ 200,  700, -50, -400],  # BB3
            [ 400,  600, -50, -500],  # BB4
            [ 500,  400, -40, -400],  # BB5
            [ 500,  500, -50, -500],  # BB6
            [ 800,  500, -40, -600],  # BB7
            [ 300,  200, -30, -400]]  # BB8
```

```python
rhs_ineq = [0, 0, 0, 0, 0, 0, 0, 0]
```

```python
lhs_eq = [[0, 0, 50, 500]] 
rhs_eq = [1]
```

```python
bnd = [(0, float("inf")), 
       (0, float("inf")), 
       (0, float("inf")), 
       (0, float("inf"))]
```

```python
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              A_eq=lhs_eq, b_eq=rhs_eq, bounds=bnd,
              method="revised simplex")
```

```python
opt
```

```python
print("Effizienz Serviceeinheit 4: {}".format(-1*opt.fun))
print("Nutzengewichte Arbeitsstunde und Materialeinsatz: {}".format(opt.x[1:]))
```

```python
effizienz_bb4 = -1 * opt.fun
print(f"Effizienz Busbahnhof 4: {effizienz_bb4:.4f} (bzw. {effizienz_bb4*100:.2f}%)")

gap = 1.0 - effizienz_bb4
v1 = opt.x[2] # Gewicht für Arbeitsstunden
v2 = opt.x[3] # Gewicht für Kosten

einsparung_arbeit = gap / v1
einsparung_kosten = gap / v2

print(f"Einsparung Arbeitsstunden: {einsparung_arbeit:.1f}h")
print(f"Einsparung Kosten: ${einsparung_kosten:.2f}")
```
