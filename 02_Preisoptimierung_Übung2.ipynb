# Preisoptimierung mit 2 Segmenten

```python
# package: scipy.optimize
# https://docs.scipy.org/doc/scipy/reference/optimize.html
# https://docs.scipy.org/doc/scipy/tutorial/optimize.html
# https://stackoverflow.com/questions/13670333/multiple-variables-in-scipys-optimize-minimize
```

```python
import numpy as np
import scipy.optimize as optimize
```

```python
# case mit 2 klassen
```

```python
# demand function and variable cost
a=40000
b=500
c=10
fixkosten_pro_segment = 50000
```

```python
# Nachfragefunktion
def demand(x):
    return a - b * x
```

```python
# Flexible Gewinnfunktion für n Segmente
def profit_function(prices):
    # Preise absteigend sortieren, um Fehler zu vermeiden (r1 > r2 > r3 ...)
    prices = sorted(prices, reverse=True) 
    
    total_profit = 0
    previous_demand = 0 # Für das erste Segment ist die vorherige Nachfrage 0
    
    for r in prices:
        current_demand = demand(r)
        # Menge in diesem Segment = Aktuelle Nachfrage minus Nachfrage der teureren Segmente
        segment_demand = current_demand - previous_demand
        
        # Gewinn für dieses Segment addieren
        total_profit += (r - c) * segment_demand
        previous_demand = current_demand
        
    return -total_profit # Negativ, da scipy.optimize "minimiert"
```

```python
# Wir testen 1 bis 5 Segmente
for n in range(1, 6):
    # Initial guess (Startwerte für den Algorithmus)
    # Wir verteilen die Startpreise grob zwischen dem Prohibitivpreis (80) und den Kosten (10)
    initial_guess = np.linspace(70, 20, n)
    
    # Optimierung durchführen
    result = optimize.minimize(profit_function, initial_guess, method="Powell")
    
    if result.success:
        brutto_gewinn = -result.fun
        netto_gewinn = brutto_gewinn - (n * fixkosten_pro_segment)
        
        print(f"--- {n} Segmente ---")
        print(f"Optimale Preise: {np.round(sorted(result.x, reverse=True), 2)}")
        print(f"Bruttogewinn: CHF {round(brutto_gewinn)}")
        print(f"Nettogewinn (abzgl. Fixkosten): CHF {round(netto_gewinn)}\n")
```

```python
# gewinnfunktion mit 2 segmenten
def f1(params):   
    r1, r2 = params
    # helper function demand to read out the demand for a set price
    def demand(x):
        return a-b*x
    return -((r1-c)*demand(r1) + (r2-c)*(demand(r2)-demand(r1))) # negative value because we need to minimize
```

```python
initial_guess = [2, 1]
result = optimize.minimize(f1, initial_guess,method="Powell")
if result.success:
    fitted_params = result.x
    print("Optimal prices: {}".format(fitted_params))
    print("Maximized revenue: {}".format(-1*result.fun))
else:
    raise ValueError(result.message)
```
